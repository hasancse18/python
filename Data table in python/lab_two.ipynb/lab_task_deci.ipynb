```python
import pandas as pd  
import numpy as np  
from pprint import pprint  

df = pd.read_csv('computer.csv',  names=['Age', 'Education', 'Income', 'Marital_Status', 'Buy_Computer'])

#print(df,"\n")
 
  
def calculate_entropy(colum_name):  
    
    atribute,count = np.unique(colum_name,return_counts = True)  
    calculate_entropy = np.sum([(-count[i]/np.sum(count))*np.log2(count[i]/np.sum(count)) for i in range(len(atribute))])  
    return calculate_entropy  
  
def IG(data,split_attribute,class_name="Buy_Computer"):  
    
    total_entropy = calculate_entropy(data[class_name])  
       
    value,count= np.unique(data[split_attribute],return_counts=True)  
 #calculating total entroy of each class
    total_entropy = np.sum([(count[i]/np.sum(count))*calculate_entropy(data.where(data[split_attribute]==value[i]).dropna()[class_name]) for i in range(len(value))])  
    #print(split_) 
    info_gain = total_entropy - total_entropy  
    
    #print(split_attribute,total_entropy)
    return info_gain  
  


def decision_tree(data,table_data,features,class_name="Buy_Computer",parent_node = None):  
  
 
    temp_var=np.unique(data[class_name])
    if  len(temp_var) <= 1:  
        return np.unique(data[class_name])[0]  
      
    elif len(data)==0:  
        return np.unique(table_data[class_name])[np.argmax(np.unique(table_data[class_name],return_counts=True)[1])]  
      
    elif len(features) ==0:  
        return parent_node  
      

    else:  
        parent_node = np.unique(data[class_name])[np.argmax(np.unique(data[class_name],return_counts=True)[1])]  

        item_values = [IG(data,feature,class_name) for feature in features] 
        best_feature_index = np.argmax(item_values)  
        best_feature = features[best_feature_index]  
          
      
        tree = {best_feature:{}}  
          

        features = [i for i in features if i != best_feature]  
          
   
          
        for value in np.unique(data[best_feature]):  
            value = value  
            sub_data = data.where(data[best_feature] == value).dropna()  
              
            subtree = decision_tree(sub_data,df,features,class_name,parent_node)  
            tree[best_feature][value] = subtree  
              
        return(tree)      

  
def train_test_split(df):  
    training_data = df.iloc[:21].reset_index(drop=True)   
    testing_data = df.iloc[21:].reset_index(drop=True)  
    return training_data,testing_data  
  

training_data = train_test_split(df)[0]

#print("training data \n",training_data)
testing_data = train_test_split(df)[1]  

#print("testing data \n",testing_data)


tree = decision_tree(training_data,training_data,training_data.columns[:-1])  
pprint(tree)
```
