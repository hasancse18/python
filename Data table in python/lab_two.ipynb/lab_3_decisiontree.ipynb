```python
import pandas as pd
import numpy as np
eps = np.finfo(float).eps
from numpy import log2 as log
import pprint




def B_tree(df,tree=None): 
    Class = df.keys()[-1]  #hold the last attribute or class class attribute of the table
    
    
    node = max_info_gain(df)#Taking the Information gain 
    
    Atribute_value = np.unique(df[node])#Create an empty dictionary to create tree    
    if tree is None:                    
        tree={}
        tree[node] = {}#We make loop to construct a tree by calling 
        #this function recursively. #In this we check if the subset is pure and stops if it is pure. 
    for value in Atribute_value:
        subtable = get_subtable(df,node,value)
        #print(subtable)
        clValue,counts = np.unique(subtable['Buy_computer'],return_counts=True)                        
        
        if len(counts)==1:#Checking purity of subset
            tree[node][value] = clValue[0]                                                    
        else:        
            tree[node][value] = B_tree(subtable) #Calling the function recursively               
    return tree

def max_info_gain(df):
    Entropy_att = []
    
    
    IG = []
    for key in df.keys()[:-1]: 
        cls_entropy=find_entropy(df)
        entropy_of_attribute=entory_of_each_attribute(df,key)
        IG.append(cls_entropy - entropy_of_attribute) #information gain
        # Calling the find_entropoy function and entory_of_each_attribute
    return df.keys()[:-1][np.argmax(IG)] 





def find_entropy(df):
    #Calculating the entropy of Buy computer class
    Class = df.keys()[-1]  
    
    entropy = 0
    values = df[Class].unique()
    
    
    for value in values:
        probability = df[Class].value_counts()[value]/len(df[Class])#calculate probability
        
        
        entropy += -probability*np.log2(probability)
    return entropy
    
    
    
    
def entory_of_each_attribute(df,attribute):
    Class = df.keys()[-1]   
    target_variables = df[Class].unique() 
   
   
    variables = df[attribute].unique()    #This gives different features in that attribute (like 18-35,36-55,<55)
    entropy2 = 0
    for variable in variables:
        entropy = 0
        for target_variable in target_variables:
            num = len(df[attribute][df[attribute]==variable][df[Class] ==target_variable]) #calculating how many "Yes" for each attribute
            den = len(df[attribute][df[attribute]==variable])#how many column of each attribute for  buy computer
            #print(df[attribute])
            #print(den)
            probability = num/(den+eps)
            entropy += -probability*log(probability+eps)
        probability2 = den/len(df)
        entropy2 += -probability2*entropy
        #print(probability)
    return abs(entropy2)    






def get_subtable(df, node,value):
    return df[df[node] == value].reset_index(drop=True)


df=pd.read_csv('computer.csv')
tree=B_tree(df)
pprint.pprint(tree)

 
```
